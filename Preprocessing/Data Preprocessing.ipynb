# User define Function for Removing outliers


```python
def remove_outlier(df , col_in):
    q1 = df[col_in].quantile(0.25)
    q3 = df[col_in].quantile(0.75)
    value = df[col_in].median()
    iqr = q3-q1
    x = list(df[col_in])
    y = []
    fence_low  = q1-1.5*iqr
    fence_high = q3+1.5*iqr
    for i in x:
        if (i < fence_low) or (i > fence_high):
            i = value
            y.append(i)
        else:
            y.append(i)
    df.drop(col_in,axis=1,inplace=True)
    df[col_in] = y
```

# User define Function for Removing Zero Values by Mean Value 


```python
def remove_zero_by_mean_value(df,col_in):
    x = df[df[col_in] != 0][col_in].median()
    df[col_in].replace(0,x,inplace=True)
```
