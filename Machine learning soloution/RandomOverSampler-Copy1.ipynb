```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.simplefilter(action='ignore', category=UserWarning)
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
df = pd.read_excel('preprocessed data.xlsx')
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_res = df.iloc[:,3:].values
y_res = df.iloc[:,2].values
```

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
os = RandomOverSampler()
X_res,y_res = os.fit_sample(X_res,y_res)
```

```python
X_res.shape
```

```python
y_res.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_res,y_res, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import Normalizer
```

```python
scaler = Normalizer().fit(X_train)
standarized_x = scaler.transform(X_train)
standarized_x_test = scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
model = RandomForestClassifier()
```

```python
model.fit(standarized_x,y_train)
y_pred = model.predict(standarized_x_test)
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
from sklearn.model_selection import cross_val_score
score = cross_val_score(model,X_res,y_res,cv=10)
```

```python
X_res = pd.DataFrame(X_res)
y_res = pd.DataFrame(y_res)
```

```python
from sklearn.model_selection import StratifiedKFold
```

```python
accuracy = []

skf = StratifiedKFold(n_splits=10, random_state= None)
skf.get_n_splits(X_res,y_res)

for train_index, test_index in skf.split(X_res,y_res):
    print("Train:",train_index, "Val",test_index)
    X1_train, X1_test = X_res.iloc[train_index],X_res.iloc[test_index]
    y1_train, y1_test = y_res.iloc[train_index],y_res.iloc[test_index]
    
    model.fit( X1_train,  y1_train)
    prediction = model.predict(X1_test)
    score = accuracy_score(prediction,y1_test)
    accuracy.append(score)
    
print(accuracy)
    
```

```python
np.array(accuracy).mean()
```

```python
from sklearn.metrics import roc_auc_score



# Define a parameters grid
param_grid = {
     'max_depth': [3, None],
     'n_estimators': [10, 20]
}

NCV = NestedCV(model=RandomForestClassifier(), params_grid=param_grid,
               outer_kfolds= 10, inner_kfolds= 10,
               cv_options={'metric':roc_auc_score, 
                           'metric_score_indicator_lower':False,
                           'randomized_search_iter':30, 
                           'predict_proba':True})
NCV.fit(X=X_res ,y=y_res)

score_k = NCV.outer_scores
```
